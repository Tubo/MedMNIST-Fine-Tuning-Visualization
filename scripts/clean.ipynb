```python
import pandas as pd
```

```python
df = pd.read_csv("../data/all_datasets_summary.csv")
df.head()
```

```python
chest = pd.read_csv("../data/chestmnist_summary.csv")
chest.head()
```

```python
set(chest.columns) - set(df.columns)
```

```python
chest_extract = df[df['dataset'] == 'chestmnist']
```

```python
chest_extract
```

```python
chest_extract['test_acc'].value_counts()
```

```python
# replace the chestmnist row with the chest df row in column test_acc, matched by dataset name and strategy name
for idx, row in chest.iterrows():
    dataset = row['dataset']
    strategy = row['strategy']
    test_acc = row['test_acc']
    
    # Find the corresponding row in df
    mask = (df['dataset'] == dataset) & (df['strategy'] == strategy)
    
    # Update the test_acc value in df
    df.loc[mask, 'test_acc'] = test_acc
```

```python
df.to_csv("../all_datasets_summary.csv")
```
